```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],  # Numerical data
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female'],  # Categorical data
    'Income': [50000, 54000, 70000, 68000, 72000]  # Numerical data
}

df = pd.DataFrame(data)
df.head()
```

```python
# Accessing categorical data
print("Categorical Data (Gender):")
df['Gender']
```

```python
# Accessing Numerical data
print("Numerical Data (Age):")
df['Age']
```

```python
# Plotting numerical data
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.bar(df['Name'], df['Age'], color='blue')
plt.title('Age of Individuals')
plt.xlabel('Name')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
plt.bar(df['Name'], df['Income'], color='green')
plt.title('Income of Individuals')
plt.xlabel('Name')
plt.ylabel('Income')

plt.tight_layout()
plt.show()
```

```python
# Plotting categorical data
gender_counts = df['Gender'].value_counts()
plt.figure()
gender_counts.plot(kind='bar', color=['red', 'blue'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
